## Applying Linear Regression on Length of Membership and Yearly amount spent 

```python
import numpy as np
X= np.array(df['Length_of_Membership']).reshape(-1,1)
y= np.array(df['Yearly_Amount_Spent']).reshape(-1,1)

#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.33, random_state=85)
```

```python
from sklearn.linear_model import LinearRegression
lm= LinearRegression()
model= lm.fit(X_train, y_train)
y_pred= lm.predict(X_test)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()

# Calculating Root mean square error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

### The above graph is fitting somewhat linearly i.e. Amount Spent is directly dependent on Length of Membership 

## Applying linear model between Time on App and Yearly amount spend

```python
X2= np.array(df['Time_on_App']).reshape(-1,1)
y2= np.array(df['Yearly_Amount_Spent']).reshape(-1,1)

#Splitting the dataset
from sklearn.model_selection import train_test_split
X2_train, X2_test, y2_train, y2_test=train_test_split(X2,y2, test_size=0.33, random_state=85)

from sklearn.linear_model import LinearRegression
lm= LinearRegression()
model= lm.fit(X2_train, y2_train)
y2_pred= lm.predict(X2_test)

plt.scatter(y2_test,y2_pred)
plt.xlabel('y2_test')
plt.ylabel('y2_pred')
plt.show()

# Calculating Root Mean Square Error
from sklearn.metrics import mean_squared_error
mean_squared_error(y2_test, y2_pred)
```

## Applying linear model between Time on Website and Yearly amount spend

```python
X3= np.array(df['Time_on_Website']).reshape(-1,1)
y3= np.array(df['Yearly_Amount_Spent']).reshape(-1,1)

#Splitting the dataset
from sklearn.model_selection import train_test_split
X3_train, X3_test, y3_train, y3_test=train_test_split(X3,y3, test_size=0.33, random_state=85)

from sklearn.linear_model import LinearRegression
lm= LinearRegression()
model= lm.fit(X3_train, y3_train)
y3_pred= lm.predict(X3_test)

plt.scatter(y3_test,y3_pred)
plt.xlabel('y3_test')
plt.ylabel('y3_pred')
plt.show()

# Calculating Root Mean Square Error
from sklearn.metrics import mean_squared_error
mean_squared_error(y3_test, y3_pred)
```

### Analysing the above graphs and their RMSE value, the company should more focus on Mobile App as more Time spent on App is generating more revenue. At the same time, company should also focus on its Website, so as to generate revenue from Website also.
